```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
left,right=plt.xlim()           #check the current axes range
print(left,right) 
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlim(1,6)  # also take tuple((1,6))  ,(2) also can take singlee value
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlim(left=3,right=7)  #change value right or left
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
bottom,top=plt.ylim()    # y lim
print(bottom,top)# (bottom ,top)
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
