```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color="r",label="square")
plt.plot(t**3,t,color="g",label="cube")

plt.grid()

plt.xlim(0,5)
print(plt.xticks([]))
print(plt.yticks([]))

plt.legend()


plt.xlabel("x-axix")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color="r",label="square")
plt.plot(t**3,t,color="g",label="cube")

plt.grid()

plt.xlim(0,20)
plt.xticks(np.arange(1,21,step=4),["A","B","C","D",'E'],color="r",size="xx-large",rotation="vertical")



plt.xlabel("x-axix")
plt.ylabel("y-axis")
plt.show()
```
