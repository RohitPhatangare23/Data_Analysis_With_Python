```python
import matplotlib.pyplot as plt
import numpy as np
t=[0,30,45,60,90]
x=[i*(np.pi/180) for i in t]
plt.plot(t,np.sin(x),marker="+",color="r",label="sine")
plt.plot(t,np.cos(x),marker="^",color="g",label="cos")


plt.xticks(t)

plt.grid()

plt.legend()



plt.title("Sine Cosine Graph")
plt.xlabel("angel")
plt.ylabel("sine and cosine value")
plt.show()
```
