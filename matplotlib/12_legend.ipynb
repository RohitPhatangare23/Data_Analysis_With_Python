```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color="r")   # or label="square"
plt.plot(t**3,t,color="b")  # or label="cube" 
plt.legend(["Square","Cube"])   #order of lable is dpend up orederof the element
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
line1,=plt.plot(t**2,t,color="r")   
line2,=plt.plot(t**3,t,color="b")   
plt.legend([line1,line2],["square","cube"])   
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
