```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([10,20,30,40],marker="+",color="r")

plt.text(1,20,"points",bbox={"ec":"y"})     #ec=edge color
plt.text(2,30,"this is point of add",bbox={"fc":"r"})   #fc=face color

plt.grid()


plt.title("Sine Cosine Graph")
plt.xlabel("angel")
plt.ylabel("sine and cosine value")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([10,20,30,40],marker="+",color="r")

plt.text(1,20,"points",bbox={"ec":"y","boxstyle":"circle"})     #ec=edge color
plt.text(2,30,"this is point of add",bbox={"fc":"r","boxstyle":"sawtooth"})   #fc=face color
#larrow , rarrow    circle,
plt.grid()


plt.title("Sine Cosine Graph")
plt.xlabel("angel")
plt.ylabel("sine and cosine value")
plt.show()


# Class	    Name	         Attrs

# Circle	circle         pad=0.3
# DArrow	darrow	       pad=0.3
# LArrow	larrow	       pad=0.3
# RArrow	rarrow	       pad=0.3
# Round	    round          pad=0.3, rounding_size=None
# Round4	   round4      pad=0.3, rounding_size=None
# Roundtooth	oundtooth  pad=0.3, tooth_size=None
# Sawtooth	   sawtooth    pad=0.3, tooth_size=None
# Square	      square   pad=0.3




```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([10,20,30,40],marker="+",color="r")


plt.annotate("(1,20)",(1,20),(0.5,25),arrowprops={},bbox={"boxstyle":"circle,pad=0.9"})   

plt.grid()


plt.title("Sine Cosine Graph")
plt.xlabel("angel")
plt.ylabel("sine and cosine value")
plt.show()
```
