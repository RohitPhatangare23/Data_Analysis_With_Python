```python
import matplotlib.pyplot as plt
import numpy as np


x=["science","commerce","arts"]
h=[200,300,500]
w=[0.5,0.3,0.9]   #width
b=[100,200,300]   #bottom
plt.bar(x,h,w,b)


plt.grid()


plt.title("Courses")
plt.xlabel("student enrooled")
plt.ylabel("student enrolld for diff courses")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x=["science","commerce","arts"]
h=[200,300,500]
# w=[0.5,0.3,0.9]   #width
# b=[100,200,300]   #bottom
plt.bar(x,h,width=-0.5,align="edge")


plt.grid()


plt.title("Courses")
plt.xlabel("student enrooled")
plt.ylabel("student enrolld for diff courses")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x=["science","commerce","arts"]
h=[200,300,500]
# w=[0.5,0.3,0.9]   #width
# b=[100,200,300]   #bottom
c=["g","yellow","b"]
plt.bar(x,h,width=0.5,color=c,linewidth=20,edgecolor="r")





plt.title("Courses")
plt.xlabel("student enrooled")
plt.ylabel("student enrolld for diff courses")
plt.show()
```
