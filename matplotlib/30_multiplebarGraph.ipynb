```python
import matplotlib.pyplot as plt
import numpy as np

w=0.3
x=["USA","india","saouth africa","america","myanmar"]

boys=[20,10,6,24,30]
girls=[100,50,30,120,180]


bar1=np.arange(len(x))
bar2=[i+w for i in bar1]

plt.bar(bar1,boys,w,label="boy")
plt.bar(bar2,girls,w,label="girl")
plt.xticks(bar1+w/2,x)
plt.legend()

plt.title("courses")
plt.xlabel("student")
plt.ylabel("courses vs student")
plt.show()
```
