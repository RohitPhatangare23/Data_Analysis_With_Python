```python
# stacked histogram

import matplotlib.pyplot as plt
import numpy as np


x=np.random.rand(100,5)
plt.hist(x,30,histtype="barstacked")   #stacked=True





plt.title("population age count")
plt.xlabel("age")
plt.ylabel("count")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x=np.random.rand(100)
y=np.sin(x)
z=np.cos(x)

plt.hist(x,bins="auto",ec="k",alpha=0.5,stacked=True)   
plt.hist(y,bins="auto",ec="k",alpha=0.5)   
plt.hist(z,bins="auto",ec="k")   



plt.title("population age count")
plt.xlabel("age")
plt.ylabel("count")
plt.show()
```
