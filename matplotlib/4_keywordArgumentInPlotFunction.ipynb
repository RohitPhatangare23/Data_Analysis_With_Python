```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8],linewidth=5,antialiased=True,c="g",dashes=[5,10])
plt.plot([8,3,4,5],[14,16,17,28],dashes=[5,10])
plt.plot([11,25,3,41],[65,56,47,38],dashes=[2,7])
plt.plot([14,24,36,43],[55,64,77,83],dashes=[3,9])
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot([1,2,3,4],[5,6,7,8],ls=":")
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot([1,2,3,4],marker="+",markeredgecolor="k",markeredgewidth=23)
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot([1,2,3,4],marker="o",markersize=10,markerfacecolor="r")
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot([10,20,30,40,50,60,70,80,90],marker="o",markersize=10,markerfacecolor="r",markevery=2)
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
