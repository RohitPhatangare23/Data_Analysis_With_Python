```python
# Anotation=a note by a way of explanation or
#           comment added to a text or dig

# annonate(s,xy,*arg,*kwargs)  s=x,y=float

# text(x,y,s,fontdict=none,withdash=false,**kwargs)
# x,y=position ,where u need to add text in plot
#s= string


import matplotlib.pyplot as plt
import numpy as np

plt.plot([10,20,30,40],marker="+",color="r")
plt.annotate("point",(2,25),size=20,color="r")





plt.grid()


plt.title("Sine Cosine Graph")
plt.xlabel("angel")
plt.ylabel("sine and cosine value")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([10,20,30,40],marker="+",color="r")
plt.annotate("point",(2,25),(0.5,25),arrowprops={"arrowstyle":"<->"})
# search arrowstyle matplotlib

plt.grid()


plt.title("Sine Cosine Graph")
plt.xlabel("angel")
plt.ylabel("sine and cosine value")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([10,20,30,40],marker="+",color="r")
plt.annotate("point",(1,20),(0,45),annotation_clip=False)
# search arrowstyle matplotlib

plt.grid()


plt.title("Sine Cosine Graph")
plt.xlabel("angel")
plt.ylabel("sine and cosine value")
plt.show()
```
