```python
# text(x,y,s,fontdict=none,withdash=false,**kwargs)
# x,y=position ,where u need to add text in plot
#s= string


import matplotlib.pyplot as plt
import numpy as np

plt.plot([10,20,30,40],marker="+",color="r")


plt.text(2.5,25,"note")
print(plt.ylim())

plt.grid()


plt.title("Sine Cosine Graph")
plt.xlabel("angel")
plt.ylabel("sine and cosine value")
plt.show()
```

```python
# text(x,y,s,fontdict=none,withdash=false,**kwargs)
# x,y=position ,where u need to add text in plot
#s= string


import matplotlib.pyplot as plt
import numpy as np

plt.plot([10,20,30,40],marker="+",color="r")

ax=plt.gca()
plt.text(0.5,0.5,"note",transform=ax.transAxes)
print(plt.ylim())

plt.grid()


plt.title("Sine Cosine Graph")
plt.xlabel("angel")
plt.ylabel("sine and cosine value")
plt.show()
```

```python
# text(x,y,s,fontdict=none,withdash=false,**kwargs)
# x,y=position ,where u need to add text in plot
#s= string


import matplotlib.pyplot as plt
import numpy as np

plt.plot([10,20,30,40],marker="+",color="r")


plt.text(2.5,25,"note",{"size":20,"color":"r"})  # also u add without dictionary
print(plt.ylim())

plt.grid()


plt.title("Sine Cosine Graph")
plt.xlabel("angel")
plt.ylabel("sine and cosine value")
plt.show()
```
