```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.title("Simple Graph")
plt.xlabel("X-axis",{"size":20,"color":"g","backgroundcolor":"k"},alpha=0.49)
plt.ylabel("Y-axis",size="xx-small",color="b",backgroundcolor="y")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.title("Simple Graph")
plt.xlabel("X-axis",{"size":20,"color":"g","backgroundcolor":"k"},alpha=0.49,style="italic")       #font style== normal,italic,oblique
plt.ylabel("Y-axis",size="xx-small",color="b",backgroundcolor="y")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.title("Simple Graph")
plt.xlabel("X-axis",{"size":20,"color":"g","backgroundcolor":"k","weight":"black"},alpha=0.49)  # fontweight or weight 
plt.ylabel("Y-axis",size=30,color="b",backgroundcolor="y",weight=200)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.title("Simple Graph")
plt.xlabel("X-axis",{"size":20,"color":"g","backgroundcolor":"k"},alpha=0.49,family="fantasy")           # fontfamily or family
plt.ylabel("Y-axis",size="30",color="b",backgroundcolor="y")
plt.show()
```
