```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color="r",label="square")   
plt.plot(t**3,t,color="b",label="cube")  
plt.plot(t,t**2,color="k",label="fourpow") 
plt.plot(t,t**3,color="g",label="fivepow") 
plt.xlim(0,4)
plt.ylim(0,4)
plt.grid()
plt.legend(loc=(1.1,0))  
plt.tight_layout()
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color="r",label="square")   
plt.plot(t**3,t,color="b",label="cube")  
plt.plot(t,t**2,color="k",label="fourpow") 
plt.plot(t,t**3,color="g",label="fivepow") 
plt.xlim(0,4)
plt.ylim(0,4)
plt.grid()
plt.legend(loc=10,bbox_to_anchor=(1,0.1,1,0.5))  
plt.tight_layout()
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show() 
```
