```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color="r",label="square")
plt.plot(t**3,t,color="g",label="cube")

plt.grid()

plt.xlim(0,5)

plt.tick_params(labelsize=20,labelcolor="r")   #labelsize="xx-large"....
plt.legend()


plt.xlabel("x-axix")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color="r",label="square")
plt.plot(t**3,t,color="g",label="cube")

plt.grid()

plt.xlim(0,5)

plt.tick_params(colors="r",width=10)   #color=tick color,colors=both
plt.legend()


plt.xlabel("x-axix")
plt.ylabel("y-axis")
plt.show()
```
