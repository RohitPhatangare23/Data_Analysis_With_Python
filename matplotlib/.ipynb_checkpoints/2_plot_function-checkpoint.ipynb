```python
import matplotlib.pyplot as plt
plt.plot([10,20,30,40])    # atleat insert  y value  (plot with only set of value) take defult value of axis
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot([10,20,30,40],"r+",markersize=12,linewidth=50)  # format (fmt) is string 
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
