```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color="r",label="square")
plt.plot(t**3,t,color="g",label="cube")

plt.grid()

plt.xlim(0,5)

plt.tick_params(axis="x",length=10)
plt.legend()


plt.xlabel("x-axix")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color="r",label="square")
plt.plot(t**3,t,color="g",label="cube")

plt.grid()

plt.xlim(0,5)

plt.tick_params(axis="x",length=10,color="r",direction="out")#  in , inout,out
plt.legend()


plt.xlabel("x-axix")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
plt.plot(t**2,t,color="r",label="square")
plt.plot(t**3,t,color="g",label="cube")

plt.grid()

plt.xlim(0,5)
plt.minorticks_on()
plt.tick_params(width=5,length=10,color="r",which="both")#major,minor,both
plt.legend()


plt.xlabel("x-axix")
plt.ylabel("y-axis")
plt.show()
```
