```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.grid()
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.grid(b=False)                     #defult b=True
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.grid(color="r")                # plt.grid()  dobar grid print karga too grid nahi hoga
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.grid(color="r")
plt.minorticks_on()            #
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.minorticks_on() 
plt.grid(b=True,color="g",which="minor")
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.minorticks_on() 
plt.grid(b=True,color="y",linewidth=20)
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.grid(axis="y")        #x,y,both
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
