```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
line1,=plt.plot(t**2,t,color="r")   
line2,=plt.plot(t**3,t,color="b")   
plt.legend([line1,line2],["square","cube"],loc=7,ncol=2)   
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.array([1,2,3,4])
line1,=plt.plot(t**2,t,color="r",label="square")   
line2,=plt.plot(t**3,t,color="b",label="cube")  
line3,=plt.plot(t**4,t,color="k",label="fourpow") 
line4,=plt.plot(t**5,t,color="g",label="fivepow") 
plt.legend(ncol=3)   
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
