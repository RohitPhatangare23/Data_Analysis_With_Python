```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8],alpha=0.25,linewidth=10)
plt.plot([10,20,30,40],linewidth=10)
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot([10,20,80,60],linewidth=10)
plt.plot([10,20,30,40],linewidth=10)
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot([10,20,80,60],linewidth=10,zorder=2)
plt.plot([10,20,30,40],linewidth=10,zorder=1)
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot([0,0],[-1,1],linewidth=10)
plt.plot([-1,1],[-1,1],linewidth=10)
plt.plot([-1,1],[1,-1],linewidth=10)
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot([0,0],[-1,1],linewidth=10,zorder=3)
plt.plot([-1,1],[-1,1],linewidth=10,zorder=2)
plt.plot([-1,1],[1,-1],linewidth=10,zorder=1)
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
