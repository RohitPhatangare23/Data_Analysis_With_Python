```python
# types of plot ===
# line plot,bar graph,histogram,scatter plot,area plot,pie chart

import matplotlib.pyplot as plt
import numpy as np

year=[2011,2014,2015,2017,2019,2021]
unemploy=[9.8,11,12,2.6,4,16]

plt.plot(year,unemploy,marker="o",linestyle="--")


plt.title("year vs unemployment rate")
plt.xlabel("year")
plt.ylabel(" unemployement rate")
plt.show()



```
