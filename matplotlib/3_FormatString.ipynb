```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8],"c*:")
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
a=np.arange(0.6,3,0.5)
plt.plot(a,"c*:",a**3,"r^--",a**2,"g+-",a**4,"k^-.")
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
